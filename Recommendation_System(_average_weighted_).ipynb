```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
credit = pd.read_csv("C:/Users/RYuK/Recommendation_system/movies_data/tmdb_5000_credits.csv")
movies = pd.read_csv("C:/Users/RYuK/Recommendation_system/movies_data/tmdb_5000_movies.csv")
```

```python
movies.head()
```

```python
# Renaming Credits movie_id to id
credit.rename(columns={'movie_id': 'id'}, inplace=True)
# Merging Credits and Movies on ID
df = credit.merge(movies, on="id")
```

```python
df.head()
```

```python
df.info()
```

```python
# Dropping unnecesssary columns
df.drop(columns=["title_x", "title_y", "homepage","production_countries", "status"], inplace=True)
```

```python
df.info()
```

# Using Weighted Average for Movie's Average Rating


```python
# Creating variable based on the formula above
R = df['vote_average']
v = df['vote_count']
m = df['vote_count'].quantile(0.70)
C = df['vote_average'].mean()
```

```python
# Applying the formula 
df['weighted_average'] = ((R*v) + (C*m)) / (v+m)
```

```python
df.head()
```

```python
# Sorting df at weighted_average
df_sort = df.sort_values('weighted_average', ascending=False)
df_sort[["original_title","popularity", "vote_average", "vote_count", "weighted_average"]].head(20)
```

```python
plt.figure(figsize=(12,6))
ax = sns.barplot(x = df_sort['weighted_average'].head(10), y = df_sort['original_title'].head(10))

plt.xlim(0, 10)
plt.ylabel("Original Movie Title", weight='bold')
plt.xlabel("Weighted Average", weight='bold')
plt.title("Top 10 Best Movies by Weighted Average", weight='bold')
plt.show()
```

```python
# Sorting df at weighted_average
df_sort_popularity = df.sort_values('popularity', ascending=False)
df_sort_popularity[["original_title", "vote_average", "vote_count", "weighted_average", "popularity"]].head(20)
```

```python
plt.figure(figsize=(12,6))
ax = sns.barplot(x = df_sort_popularity['popularity'].head(10), y = df_sort_popularity['original_title'].head(10))

plt.ylabel("Original Movie Title", weight='bold')
plt.xlabel("Popularity", weight='bold')
plt.title("Top 10 Best Movies by Popularity", weight='bold')
plt.show()
```

# Recommendation based on both Popularity and Weighted Average (Priority given 50% each)

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
```

```python
df[['normalized_weighted_average', 'normalized_popularity']] = scale.fit_transform(df[['weighted_average', 'popularity']])
```

```python
df.head()
```

```python
# Creating Score with both 50% of popularity and 50% of weighted_average
df['score'] = df['normalized_popularity'] *0.5 + df['normalized_weighted_average']*0.5
df_score_sort = df.sort_values(['score'], ascending=False)
```

```python
df_score_sort[["original_title", "normalized_weighted_average", "normalized_popularity", "score"]].head(10)
```

```python
plt.figure(figsize=(12,6))
ax = sns.barplot(x = df_score_sort['score'].head(10), y = df_score_sort['original_title'].head(10))

plt.ylabel("Original Movie Title", weight='bold')
plt.xlabel("Score", weight='bold')
plt.title("Top 10 Best Movies by Score", weight='bold')
plt.show()
```
